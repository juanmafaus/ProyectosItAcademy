El esquema creado (transaccions) tiene dos tablas: transaction y company, las mismas tienen una relacion de uno a muchos (de company a transaction) a traves de la primary key de company (id)
y de la foreign key presente en la tabla transaction (company_id) corresponden al modelo de entidad relacional.
La tabla transaction tiene las siguientes variables(las cuales representan columnas):
id: es la primary key de esta tabla y su data type es varchar y su limite son 255 caracteres
credit_card_id: su data type es varchar y su limite son 15 caracteres
company_id: su data type es varchar y su limite son 20 caracteres (es la foreign key de la tabla company)
user_id: su data type es int (integer)
lat: su data type es float
longitude: su data type es float
timestamp: su data type es timestamp
amount: su data type es decimal y tiene 10 como numero maximo de digitos y 2 como numero maximo de digitos a la derecha del punto
declined: su data type es tinyint y su limite es 1
La tabla company tiene las siguientes variables( las cuales representan columnas):
id: su data type es varchar , su limite son 15 caracteres y es la primary key de esta tabla
company_name: su data type es varchar y su limite son 255 caracteres
phone: su data type es varchar y su limite son 15 caracteres
email: su data type es varchar y su limite son 100 caracteres
country: su data type es varchar y su limite son 100 caracteres
