Mostra el preu del producte més car comprat per cada usuari/ària.
Visualitza la distribució geogràfica dels usuaris/es.
En aquesta activitat, serà necessari que realitzis els ajustos necessaris en cada gràfic per a millorar la llegibilitat i comprensió. En el compliment d'aquesta tasca, s'espera que avaluïs acuradament quines variables són rellevants per a transmetre la informació requerida de manera efectiva.



Codigo DAX de la medida para obtener el producto mas caro que ha comprado cada usuario:

``
ProdMasCaro = 
CALCULATE(
    MAX('S4 products'[price - Copy]),
    'S4 transactions'[declined] = 0,
    CROSSFILTER('S4 product_transaction'[idProducto], 'S4 products'[id], Both)
)
``

Codigo DAX para obtener las compras por usuario:

``
comprasXusuario = CALCULATE(COUNT('S4 product_transaction'[idProducto]), 'S4 transactions'[declined] = 0)
``
Codigo DAX para obtener dos columnas a traves de un condicional que separe la media de monto por usario en mayores a 150 o menores o iguales que 150:

``
Mayor150 = IF('S4 users'[mediaMontoUsuario] > 150, "Mayor que 150", "Menor o igual que 150")
``

Codigo DAX para obtener la media de monto por usuario:
